```python
import pandas as pd
data = pd.read_csv("CONVSOLN_all_results.csv", index_col=0)
```

```python
import ast
res = ast.literal_eval(data["PATIENT"][0])
```

```python
PATIENTS = []
OARS = []
DICES = []
HDS = []
ASDS = []
EVALS = []

for i in range(len(data)):
    PATIENTS += ast.literal_eval(data["PATIENT"][i])
    OARS += ast.literal_eval(data["OAR"][i])
    DICES += ast.literal_eval(data["DICE"][i])
    HDS += ast.literal_eval(data["95HD"][i])
    ASDS += ast.literal_eval(data["ASD"][i])
    EVALS += ast.literal_eval(data["EVAL"][i])
```

```python
data_ = pd.DataFrame.from_dict({"PATIENT_ID":PATIENTS, "OAR":OARS, "DICE":DICES, "95HD":HDS, "ASD":ASDS, "EVAL":EVALS})
```

```python
data_["EVAL"]=data_["EVAL"]*100
```

```python
data_.to_csv("CONVSOLN_all_results_updated.csv")
```
